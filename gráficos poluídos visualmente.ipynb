```python
import pandas as pd
import matplotlib.pyplot as plt


dados = pd.read_csv("C:\\Users\msdof\\Downloads\\projeto parte 1 visualizacao dados\\BRA.csv")
dados
```

```python
# Renomeando as colunas
dados.rename(columns={'Season': 'temporada',
                      'Date': 'Data',
                    'Home': 'mandante', 
                    'Away': 'visitante', 
                    'HG': 'gols do mandante', 
                    'AG': 'gols do visitante',
                    'PSCH': 'odd do mandante',
                    'PSCA': 'odd do visitante',
                    'PSCD': 'odd do empate',
                    'Res': 'resultado'}, inplace=True)

# Filtrar os dados para a temporada de 2023
temporada_2023 = dados[dados['temporada'] == 2023]
temporada_2023
```

```python
# Filtrando as colunas desejadas
df = temporada_2023.loc[:, ['Data','mandante', 'visitante', 'gols do mandante', 'gols do visitante', 'resultado', 'odd do mandante', 'odd do visitante', 'odd do empate']]
df
```

```python
df['placar'] = df['gols do mandante'].astype(str) + '-' + df['gols do visitante'].astype(str)
df
# Contando a frequência de cada placar
placar_contagem = df['placar'].value_counts()
```

```python
# Cria o gráfico
plt.figure(figsize=(10, 6))
plt.plot(df['Data'], df['gols do mandante'], label='Gols do mandante')
plt.plot(df['Data'], df['gols do visitante'], label='Gols do visitante')
plt.xlabel('Data')
plt.ylabel('Gols')
plt.title('Gols ao longo da temporada 2023')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['Data'] = pd.to_datetime(df['Data'])  # Convertendo a coluna Data para datetime
df.set_index('Data', inplace=True)  # Definindo a coluna Data como índice

# Filtrando apenas os jogos do Flamengo RJ como mandante
flamengo_mandante = df[df['mandante'] == 'Flamengo RJ']

# Agrupando por mês e somando os gols
gols_por_mes = flamengo_mandante.resample('M').sum()

# Criando o gráfico de barras
plt.figure(figsize=(10, 6))
gols_por_mes[['gols do mandante', 'gols do visitante']].plot(kind='bar', stacked=False)

# Configurações do gráfico
plt.xlabel('Mês')
plt.ylabel('Total de Gols')
plt.title('Total de Gols do Flamengo RJ como Mandante por Mês')
plt.xticks(rotation=45)
plt.legend(['Gols do Mandante', 'Gols do Visitante'])
plt.tight_layout()

# Exibindo o gráfico
plt.show()
```
