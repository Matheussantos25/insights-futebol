```python
import pandas as pd
import matplotlib.pyplot as plt


dados = pd.read_csv("C:\\Users\msdof\\Downloads\\projeto parte 1 visualizacao dados\\BRA.csv")
dados
```

```python
# Renomeando as colunas
dados.rename(columns={'Season': 'temporada',
                      'Date': 'Data',
                    'Home': 'mandante', 
                    'Away': 'visitante', 
                    'HG': 'gols do mandante', 
                    'AG': 'gols do visitante',
                    'PSCH': 'odd do mandante',
                    'PSCA': 'odd do visitante',
                    'PSCD': 'odd do empate',
                    'Res': 'resultado'}, inplace=True)

# Filtrar os dados para a temporada de 2023
temporada_2023 = dados[dados['temporada'] == 2023]
temporada_2023


```

```python
# Filtrando as colunas desejadas
df = temporada_2023.loc[:, ['Data','mandante', 'visitante', 'gols do mandante', 'gols do visitante', 'resultado', 'odd do mandante', 'odd do visitante', 'odd do empate']]
df
```

```python
df['placar'] = df['gols do mandante'].astype(str) + '-' + df['gols do visitante'].astype(str)
df
# Contando a frequência de cada placar
placar_contagem = df['placar'].value_counts()
```

```python
# Plotando um gráfico de barras com os placares mais comuns
plt.figure(figsize=(10, 6))  # Define o tamanho do gráfico
placar_contagem.plot(kind='bar')
plt.xlabel('Placar')
plt.ylabel('Frequência')
plt.title('Placares Mais Comuns')
plt.xticks(rotation=45)  # Rotaciona os labels do eixo x para melhor visualização
plt.show()
```

```python
# Filtrando jogos do Flamengo RJ como mandante
flamengo_mandante = df[df['mandante'] == 'Flamengo RJ']

flamengo_mandante['Data'] = pd.to_datetime(flamengo_mandante['Data'])

# Ordenando por data
flamengo_mandante = flamengo_mandante.sort_values('Data')

# Calculando estatísticas ao longo das datas
df_estatisticas = pd.DataFrame({
    'Data': flamengo_mandante['Data'],
    'Gols': flamengo_mandante['gols do mandante']
})

# Calculando moda móvel, média móvel e mediana móvel
window_size = 5  # Tamanho da janela para cálculo da média móvel, mediana móvel e moda móvel

# Função para calcular a moda móvel
def moda_movel(serie):
    return serie.rolling(window=window_size).apply(lambda x: x.mode()[0] if not x.mode().empty else x.mean())

df_estatisticas['Moda'] = moda_movel(df_estatisticas['Gols'])
df_estatisticas['Média'] = df_estatisticas['Gols'].rolling(window=window_size).mean()
df_estatisticas['Mediana'] = df_estatisticas['Gols'].rolling(window=window_size).median()

# Gerando o gráfico de linha
fig, ax = plt.subplots()

# Plotando moda, média e mediana dos gols ao longo do tempo
ax.plot(df_estatisticas['Data'], df_estatisticas['Moda'], marker='o', linestyle='-', color='r', label='Moda')
ax.plot(df_estatisticas['Data'], df_estatisticas['Média'], marker='o', linestyle='-', color='g', label='Média')
ax.plot(df_estatisticas['Data'], df_estatisticas['Mediana'], marker='o', linestyle='-', color='b', label='Mediana')

# Adicionando rótulos e título
ax.set_xlabel('Data')
ax.set_ylabel('Gols')
ax.set_title('Moda, Média e Mediana dos Gols do Flamengo RJ como Mandante ao Longo do Tempo')
ax.legend()

# Adicionando uma grade
plt.grid(True)

# Exibindo o gráfico
plt.show()
```

```python
# Calcular a média de gols em casa (HG) e fora (AG) por time
avg_goals_home = temporada_2023.groupby('mandante')['gols do mandante'].mean().reset_index()
avg_goals_away = temporada_2023.groupby('visitante')['gols do visitante'].mean().reset_index()

# Renomear as colunas para facilitar a combinação dos datasets
avg_goals_home.columns = ['Team', 'Avg_Home_Goals']
avg_goals_away.columns = ['Team', 'Avg_Away_Goals']

# Combinar as médias de gols em casa e fora em um único dataframe
avg_goals = pd.merge(avg_goals_home, avg_goals_away, on='Team')

# Calcular a média total de gols por time
avg_goals['Avg_Total_Goals'] = (avg_goals['Avg_Home_Goals'] + avg_goals['Avg_Away_Goals']) / 2

# Ordenar por média total de gols
avg_goals = avg_goals.sort_values(by='Avg_Total_Goals', ascending=False)

# Visualizar os dados calculados
avg_goals.head()
```

```python
# Plotar a média de gols por time para a temporada de 2023
plt.figure(figsize=(12, 8))

# Configurar as barras
bar_width = 0.25
teams = avg_goals['Team']
index = range(len(teams))

# Barras para gols em casa
plt.bar(index, avg_goals['Avg_Home_Goals'], bar_width, label='Média de Gols em Casa')

# Barras para gols fora
plt.bar([i + bar_width for i in index], avg_goals['Avg_Away_Goals'], bar_width, label='Média de Gols Fora')

# Barras para média total de gols
plt.bar([i + 2 * bar_width for i in index], avg_goals['Avg_Total_Goals'], bar_width, label='Média Total de Gols')

# Adicionar títulos e rótulos
plt.xlabel('Times')
plt.ylabel('Média de Gols')
plt.title('Média de Gols por Time na Temporada de 2023')
plt.xticks([i + bar_width for i in index], teams, rotation=90)
plt.legend()

# Exibir o gráfico
plt.tight_layout()
plt.show()
```
